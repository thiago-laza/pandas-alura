#### Qual o valor medio de aluguel por tipo de imovel?

Removendo os imoveis comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
       'Prédio Inteiro','Loja/Salão',
       'Galpão/Depósito/Armazém', 'Casa Comercial'
       'Terreno Padrão',
       'Loja Shopping/ Ct Comercial','Box/Garagem','Chácara',
       'Loteamento/Condomínio','Sítio',
       'Pousada/Chalé','Hotel', 'Indústria']
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df_residenciais = dados.query('@imoveis_comerciais not in Tipo')
df_residenciais.head()
```

```python
df_residenciais.Tipo.unique()
```

```python
df_preco_tipo = df_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh',figsize=(14,10),color='purple')
```

## Aula 04

### Manipulando os dados

Criando colunas numericas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

Criando colunas categoricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quartos e ' + dados['Vagas'].astype(str) + ' de garagem'
dados.head()
```

```python
dados['Possui_suite(s)'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Nao')
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
